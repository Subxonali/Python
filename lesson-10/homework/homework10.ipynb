```python
# Homework 1: ToDo List Application (with CLI)
class Task:
    def __init__(self, title, description, date, status="Incomplete"):
        self.title = title
        self.description = description
        self.date = date
        self.status = status

    def mark_complete(self):
        self.status = "Complete"

    def display(self):
        return f"Title: {self.title}\nDescription: {self.description}\nDue Date: {self.date}\nStatus: {self.status}\n"


class ToDoList:
    def __init__(self):
        self.tasks = []

    def add_task(self, task):
        self.tasks.append(task)

    def mark_task_complete(self, title):
        for task in self.tasks:
            if task.title == title:
                task.mark_complete()
                return True
        return False

    def list_all_tasks(self):
        return [task.display() for task in self.tasks]

    def display_incomplete_tasks(self):
        return [task.display() for task in self.tasks if task.status == "Incomplete"]


# Homework 2: Simple Blog System
class Post:
    def __init__(self, title, content, author):
        self.title = title
        self.content = content
        self.author = author

    def display(self):
        return f"Title: {self.title}\nAuthor: {self.author}\nContent: {self.content}\n"


class Blog:
    def __init__(self):
        self.posts = []

    def add_post(self, post):
        self.posts.append(post)

    def list_all_posts(self):
        return [post.display() for post in self.posts]

    def display_posts_by_author(self, author):
        return [post.display() for post in self.posts if post.author == author]

    def delete_post(self, title):
        for i, post in enumerate(self.posts):
            if post.title == title:
                del self.posts[i]
                return True
        return False

    def edit_post(self, title, new_title=None, new_content=None):
        for post in self.posts:
            if post.title == title:
                if new_title:
                    post.title = new_title
                if new_content:
                    post.content = new_content
                return True
        return False

    def display_latest_posts(self, count=3):
        return [post.display() for post in self.posts[-count:]]


# Homework 3: Simple Banking System
class Account:
    def __init__(self, acc_number, holder, balance=0.0):
        self.acc_number = acc_number
        self.holder = holder
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, amount):
        if self.balance >= amount:
            self.balance -= amount
            return True
        return False

    def display(self):
        return f"Account: {self.acc_number}, Holder: {self.holder}, Balance: ${self.balance:.2f}"

    def transfer(self, other_account, amount):
        if self.withdraw(amount):
            other_account.deposit(amount)
            return True
        return False


class Bank:
    def __init__(self):
        self.accounts = {}

    def add_account(self, acc):
        self.accounts[acc.acc_number] = acc

    def get_account(self, acc_number):
        return self.accounts.get(acc_number)

    def deposit(self, acc_number, amount):
        acc = self.get_account(acc_number)
        if acc:
            acc.deposit(amount)
            return True
        return False

    def withdraw(self, acc_number, amount):
        acc = self.get_account(acc_number)
        if acc:
            return acc.withdraw(amount)
        return False

    def transfer(self, from_acc, to_acc, amount):
        acc_from = self.get_account(from_acc)
        acc_to = self.get_account(to_acc)
        if acc_from and acc_to:
            return acc_from.transfer(acc_to, amount)
        return False

    def display_account(self, acc_number):
        acc = self.get_account(acc_number)
        if acc:
            return acc.display()
        return "Account not found"

```
